## TRADITIONAL LINGUISTIC FEATURE

### LEXICAL RICHNESS

```python
cttr, brunet, std_entropy, pidensity = linguisticFeature.lexical_richness(test_transcript, lang="en")
```

```python
print(f"corrected TTR: {cttr}, burnet: {brunet}, std_entropy: {std_entropy}, pidensity: {pidensity}")
```

### PART-OF-SPEECH tag, POLARITY, SUBJECTIVITY

```python
pos_tagged_data, polarivity, subjectivity = linguisticFeature.pos_polarity_subjectivity(test_transcript, lang="en")
```

```python
print(f"pos tag: {pos_tagged_data}")
print(f"polarivity: {polarivity}")
print(f"subjectivity: {subjectivity}")
```

```python
tags_feature = linguisticFeature.tag_count(pos_tagged_data)
```

```python
print(f"tags_feature: {tags_feature}")
```

```python
# Tag rate
rates = linguisticFeature.evaluate_pos_rate(tags_feature)
```

```python
print(rates)
```

### DISFLUENCY COUNT

```python
disfluency_count = linguisticFeature.count_disfluency(test_transcript, lang="en")
```

```python
print(disfluency_count)
```

### Readability

```python
dale_chall, flesch, coleman_liau_index, r_time, syllables = linguisticFeature.evaluate_readability(test_transcript)
```

```python
print(f"dale_chall: {dale_chall}, flesch: {flesch}, coleman_liau_index: {coleman_liau_index}, r_time: {r_time}, syllables: {syllables}")
```

### Deixis evaluation

```python
person_rate, spatial_rate, temporal_rate = linguisticFeature.evaluate_deixis(test_transcript, lang="en")
```

```python
print(f"person_rate: {person_rate}, spatial_rate: {spatial_rate}, temporal_rate: {temporal_rate}")
```
